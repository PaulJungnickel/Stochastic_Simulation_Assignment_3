```python
from CircleParticleSim import *
import numpy as np
import numpy.random as rand
```

```python
rand.seed(42)
cooling_steps = 1000
markov_len = 2000
for i in range(10):
    sim = CircleParticleSim(16, steps=markov_len * cooling_steps, seed=rand.randint(0,2**31-1),
                 cooling_schedule = basic_cooling_schedule,
                 step_size_schedule = random_step_size_schedule,
                 random_step_likelihood=0.2,
                 initial_temperature=1
                 )
    
    sim.run_simulation(cooling_steps, markov_len)
    print(sim.E/2)
    sim.plot_positions()
```
